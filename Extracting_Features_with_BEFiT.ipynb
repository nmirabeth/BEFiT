### Load the model

```python
from transformers import BeitForImageClassification

model = BeitForImageClassification.from_pretrained('trainer') 
```

```python
# Take just the trained enconder
model_encoder=model.beit
```

```python
from transformers import BeitFeatureExtractor
from PIL import Image

# Load the image
img_path='//path/example.jpg'

im=Image.open(img_path)

# Data pre-processing
feature_extractor = BeitFeatureExtractor.from_pretrained('microsoft/beit-base-patch16-224-pt22k-ft22k')
encoding = feature_extractor(images=im, return_tensors="pt")
encoding.keys()

pixel_values = encoding['pixel_values']

# Pass the input through the model to obtain the embedded vector
outputs = model_encoder(pixel_values)
```
